```python
from sklearn.datasets import make_classification
import numpy as np

X,y = make_classification(n_samples=100,n_redundant=0, n_features=2 ,n_informative=1, random_state=41, n_clusters_per_class=1 , n_classes=2 ,
                          class_sep=10,hypercube=False)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c=y , cmap='rainbow' ,s=100)
```

```python
def perceptron(X,y):
    X= np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr=0.1

    for i in range(1000):
        j= np.random.randint(0,100) #selecting a random sample by selecting a random integder between 0 and 100 via index
        y_hat = step(np.dot(X[j],weights)) #Here dot product of the weights and the sample(i.e [x1,x2,x3],[w1,w2,w3] that will be x1w1+x2w2+x3w3 ) is calculated  
                                           #then passed to the step function to get the predicted value_)
        weights = weights + lr*(y[j]-y_hat)*X[j] #Here the weights are updated by adding the product of learning rate, error and the sample to the weights

    return weights[0],weights[1:]
```

```python
def step(z):
    if z > 0:
        return 1
    else:
        return 0
```

```python
intercept_,coef_ = perceptron(X, y)  # perceptron() returns (w0, [w1,w2]); store them separately for plotting/prediction.
```

```python
m = (coef_[0] / coef_[1])
b = (intercept_ / coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

```python
from sklearn.linear_model import LogisticRegression
lor = LogisticRegression()
lor.fit(X,y)
```

```python
m = -(lor.coef_[0][0]/lor.coef_[0][1])
b = -(lor.intercept_/lor.coef_[0][1])
```

```python
x_input1 = np.linspace(-3,3,100)
y_input1 = m*x_input + b
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.plot(x_input1,y_input1,color='black',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
